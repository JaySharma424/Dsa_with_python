```python
# Write a function that takes a list of integers and returns the sum of the first and last elements.
# If the list is empty, return 0. If the list has only one element, return that element multiplied by 2.
# Example Input :- [10, 20, 30, 40]
# Example Output:- 50

def sum_of_first_last(array):
  n = len(array)
  left, right = 0, n - 1
  if len(array) == 0:
    return 0
  if len(array) == 1:
    return array[0] * 2
  for i in range(n):
    sum = array[left] + array[right]
  return sum

array = eval(input())
sum_of_first_last(array)
```

```python
'''Given a string, write a function to reverse it without using Python's built-in reversed() function or extended slicing [::-1].
Example Input
"hello"
Example Output
"olleh"
'''
def reverse_string(text):
  for i in range(len(text)-1 , -1 , -1):
    print(text[i], end= '')

text = input()
reverse_string(text)
```

```python
'''
Write a function that takes a list of integers and moves all zeros to the end of the list while maintaining
the relative order of the non-zero elements. Perform this in-place (do not create a new list).
Example Input
[0, 1, 0, 3, 12]
Example Output
[1, 3, 12, 0, 0]
'''

def move_zero(arr):
  n = len(arr)
  left = 0

  for right in range(n):
    if arr[right] != 0:
      arr[left] , arr[right] = arr[right], arr[left]
      left += 1
  return arr
arr = eval(input())
move_zero(arr)
```

```python
''
Write a function to check if a given string is a palindrome. A palindrome reads the same forwards and backwards.
Ignore case and non-alphanumeric characters.
Example Input
"A man, a plan, a canal: Panama"
Example Output
True
'''
def is_palindrome(text):
  left = 0
  right = len(text)-1

  while left < right:

    if not text[left].isalnum():
      left += 1
    elif not text[right].isalnum():
      right -= 1
    else:
      if text[left].lower() != text[right].lower():
        return False
      left += 1
      right -= 1
  return True

  print(text)


text = str(input())
is_palindrome(text)
```

```python
'''
Problem Statement: Reverse each word without changing order

Example Input: "hello world python"

Example Output: "olleh dlrow nohtyp"
'''
def reverse_word(text):
  list1 = list(text.split())
  # print(list1)
  for i in range(len(list1)):
    print(list1[i][::-1], end=' ')


reverse_word("hello world python")
```

```python
'''
Problem Statement: Count frequency of each character
Example Input: "banana"
Example Output Character Frequency
b = 1
a = 3
n = 2
'''

def frequency_count(text):
  fre_count = {}
  for char in text:
    fre_count[char] = fre_count.get(char,  0) + 1

  for key, value in fre_count.items():
    # key =
    print(key, " = " , value)

frequency_count("banana")
```

```python
# Problem Statement: Find second largest number in list
def second_largest(arr):
  max1 = 0
  max2 = 0
  for num in arr:
    if num > max1:
      max2 = max1
      max1 = num
    else:
      max2 = num
  return min(max1, max2)

second_largest([69, 45, 67, 23, 50, 70])
```
