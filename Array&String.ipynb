```python
# Write a function that takes a list of integers and returns the sum of the first and last elements.
# If the list is empty, return 0. If the list has only one element, return that element multiplied by 2.
# Example Input :- [10, 20, 30, 40]
# Example Output:- 50

def sum_of_first_last(array):
  n = len(array)
  left, right = 0, n - 1
  if len(array) == 0:
    return 0
  if len(array) == 1:
    return array[0] * 2
  for i in range(n):
    sum = array[left] + array[right]
  return sum

array = eval(input())
sum_of_first_last(array)
```

```python
'''Given a string, write a function to reverse it without using Python's built-in reversed() function or extended slicing [::-1].
Example Input
"hello"
Example Output
"olleh"
'''
def reverse_string(text):
  for i in range(len(text)-1 , -1 , -1):
    print(text[i], end= '')

text = input()
reverse_string(text)
```

```python
'''
Write a function that takes a list of integers and moves all zeros to the end of the list while maintaining
the relative order of the non-zero elements. Perform this in-place (do not create a new list).
Example Input
[0, 1, 0, 3, 12]
Example Output
[1, 3, 12, 0, 0]
'''

def move_zero(arr):
  n = len(arr)
  left = 0

  for right in range(n):
    if arr[right] != 0:
      arr[left] , arr[right] = arr[right], arr[left]
      left += 1
  return arr
arr = eval(input())
move_zero(arr)
```

```python
'''
Write a function to check if a given string is a palindrome. A palindrome reads the same forwards and backwards.
Ignore case and non-alphanumeric characters.
Example Input
"A man, a plan, a canal: Panama"
Example Output
True
'''
def is_palindrome(text):
  left = 0
  right = len(text)-1

  while left < right:

    if not text[left].isalnum():
      left += 1
    elif not text[right].isalnum():
      right -= 1
    else:
      if text[left].lower() != text[right].lower():
        return False
      left += 1
      right -= 1
  return True

  print(text)


text = str(input())
is_palindrome(text)
```

```python
'''
Problem Statement: Reverse each word without changing order

Example Input: "hello world python"

Example Output: "olleh dlrow nohtyp"
'''
def reverse_word(text):
  list1 = list(text.split())
  # print(list1)
  for i in range(len(list1)):
    print(list1[i][::-1], end=' ')


reverse_word("hello world python")
```

```python
# Problem Statement: Find second largest number in list
def second_largest(arr):
  max1 = 0
  max2 = 0
  for num in arr:
    if num > max1:
      max2 = max1
      max1 = num
    else:
      max2 = num
  return min(max1, max2)

second_largest([69, 45, 67, 23, 50, 70])
```

```python
# Separate Positives and Negatives Problem: Given an integer array, move all negative numbers to the left side
# and all positive numbers to the right side in-place. The order of the numbers does not matter.
# Input: [-1, 2, -3, 4, 5, -6] Output: [-1, -6, -3, 4, 5, 2] (or any other valid arrangement)

def shift_negative_positive(arr):
  left , right = 0, len(arr) - 1
  while left < right :
    while left < right and arr[left] < 0 :
      left += 1
    while left < right and arr[right] > 0:
      right -= 1
    if left < right:
      arr[left] , arr[right] = arr[right] , arr[left]
      left += 1
      right -= 1
  return arr


shift_negative_positive([-1, 2, -3, 4, 5, -6])
```

```python
# Valid Palindrome II (One Deletion allowed) Problem: Given a string s, return True if the string can be a palindrome
# after deleting at most one character from it.Input: "abca" Output: True (You could delete the 'c' to get "aba")
# Students Hint: Start like a normal valid palindrome using two pointers. If you hit a mismatch, you have two choices:
# skip the left character OR skip the right character.If either resulting substring is a palindrome, return True!

def palindrome_with_deletion(s):
  def is_palindrome(s,left, right):
    while left < right:
      if s[left].lower() != s[right].lower():
        return False
      left += 1
      right -= 1
    return True
  left , right = 0, len(s) -1

  while left < right:
    if s[left].lower() != s[right].lower():
      return is_palindrome(s, left + 1, right) or is_palindrome(s, left, right -1)
    left += 1
    right -= 1
  return True

# Test cases
print(palindrome_with_deletion("abca"))   # True  → delete 'c' → "aba"
print(palindrome_with_deletion("racecar"))# True  → already a palindrome
print(palindrome_with_deletion("Nitish"))
```

```python
# Container With Most Water Problem: You are given an integer array heights where each element represents the height of a vertical line on a graph.
# Find two lines that together with the x-axis form a container, such that the container contains the most water.
# Return the maximum area of water it can store.
# Input: [1, 8, 6, 2, 5, 4, 8, 3, 7] Output: 49

def most_water(arr):

  left, right = 0, len(arr) - 1

  max_area = 0

  while left < right:

    min_piller = min(arr[left], arr[right])
    length = right - left
    curr_area = min_piller * length
    max_area = max(max_area, curr_area)
    if min_piller == arr[left]:
      left += 1
    else:
      right -= 1
  return max_area

most_water([1, 8, 6, 2, 5, 4, 8, 3, 7])
```
