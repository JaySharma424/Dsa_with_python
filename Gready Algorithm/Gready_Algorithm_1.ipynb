### **Greedy algorithms**
Greedy algorithms are a class of problem-solving strategies that make the locally optimal choice at each distinct stage with the hope of finding a global optimum. Instead of exhaustively searching all possible paths or computational combinations, a greedy algorithm aggressively selects the best immediate option available, never looking back or reconsidering its previous choices.

This approach fundamentally assumes that local optimums accumulate into a global optimum. For many problems, such as finding the shortest path in a graph with non-negative weights or building a minimum spanning tree, this mathematical assumption holds true perfectly. When it does hold true, greedy algorithms offer blazing fast, highly scalable solutions.

However, the greedy paradigm is not a universal solver. Its primary purpose is to deliver highly performant, often linear or log-linear time complexity solutions to optimization problems where the optimal substructure and greedy choice property are strictly satisfied. When these properties fail, greedy algorithms produce suboptimal approximations rather than definitive correct answers.

```python
# Define the function to find the minimum coins needed using a greedy approach
# coins = [1, 2, 5, 10] amount = 18 OP --> 4 (minimum coins needed)

def find_min_coin(coins, amount) :
  coins.sort(reverse = True)
  min_coin = 0
  for coin in coins :

    if coin <= 0 :
      continue
    count = amount // coin

    min_coin += count

    amount = amount % coin
  return min_coin if amount == 0 else -1

print(find_min_coin([1, 2, 5, 10], 18))
print(find_min_coin([1, 5, 10, 25], 36))
print(find_min_coin([1, 2, 5], 11))
print(find_min_coin([2,5,10,25,50],100))
```

You are given $N$ activities, each with a specific start time and finish time. An individual or a single machine can only work on a single activity at a time. Two activities cannot overlap, meaning you cannot start a new activity before the current one finishes.Your goal is to find the maximum number of activities that can be executed successfully by a single person or machine.

```python
# starts = [1, 3, 0, 5, 8, 5]
# finishes = [2, 4, 6, 7, 9, 9]
# Output: Sequence of valid intervals [(1, 2), (3, 4), (5, 7), (8, 9)]

def find_maximum_activites(start, end) :
  activites = list(zip(start, end))

  activites.sort(key = lambda x : x[1])

  last_start, last_end = activites[0][0], activites[0][1]

  result = [(last_start, last_end)]
  for start, end in activites[1:] :
    if start >= last_end :
      result.append((start, end))
      last_start, last_end = start, end
    else:
      pass
  print(activites)
  return result

print(find_maximum_activites([1, 3, 0, 5, 8, 5], [2, 4, 6, 7, 9, 9]))
print(find_maximum_activites([10, 12, 20], [20, 25, 30]))
print(find_maximum_activites([1, 2, 3], [5, 5, 5]))
```

Imagine you are in a spice market with a jar that can hold only 50 kg.

Different spices have different values per kg, such as gold dust, saffron, and black pepper.

To earn the maximum profit, you first choose the spice with the highest value per kg.

You fill the jar with that spice as much as possible before moving to the next valuable 1.

If only a little space remains, you take just a fraction of the next spice to fill the jar completely.

The Fractional Knapsack Algorithm works in the same way by selecting items based on maximum value and allowing fractions when needed.

```python
def frictional_knapsack(list, capacity) :
  # sort by friction
  list.sort(key= lambda x:x[0]//x[1], reverse = True)

  profit = 0
  for value, weight in list :
    if capacity >= weight :
      capacity -= weight
      profit += value
    else:
      frictional_value = (value // weight) * capacity
      profit += frictional_value

  return profit
print(frictional_knapsack([[120,30], [60, 10], [100, 20]], 50))
print(frictional_knapsack([[50, 10], [40, 20]], 100))
print(frictional_knapsack([[50, 10], [40, 20]], 0))
print(frictional_knapsack([[5, 10], [4, 8]], 10))
```

```python
# Problem Statement: Assume you are an awesome parent and want to give your child some cookiess.
# Each child has a specific greed factor representing the minimum size of a cookies they will be content with.
# Each cookies has a specific physical size. Maximize the total number of perfectly content child and return the numeric count.
# Example Input: Greed array [1, 2, 3], cookies array [1, 1]
# Example Output: 1

def greed_factor(greed, cookies) :
  greed.sort()
  cookies.sort()


  child = 0
  cookie = 0

  while child < len(greed) and cookie < len(cookies) :

    if greed[child] <= cookies[cookie] :
      child += 1
    cookie += 1

  return child




print(greed_factor([1, 2, 3], [1, 2, 3]))
print(greed_factor([1, 2, 3], [1, 1]))
print(greed_factor([1, 2],[3]))
print(greed_factor([5, 6, 7], [1, 2, 3]))
print(greed_factor([1], [1, 2, 3]))
print(greed_factor([1, 2], []))
```

```python
# Problem Statement: You are assigned to securely load boxes onto a truck. You are given a 2D array of box types,
# where each type designates the quantity of boxes available and the units inside each box. You are given a strict
# truck size defining the maximum boxes it can hold. Return the absolute maximum units safely loaded.
# Example Input: Box types [[1, 3], [2, 2], [3, 1]], Truck size 4.
# Example Output: Eight

def greedy_load(boxes, truck) :
  boxes.sort(key= lambda x:x[1], reverse = True)

  max_boxes = 0

  for quantity, value in boxes :

    if quantity <= truck :
      truck -= quantity
      max_boxes += value * quantity
    else:

      boxes_to_take = min(quantity, truck)
      max_boxes += boxes_to_take * value
      truck -= boxes_to_take
    if truck == 0 :
      break


  return max_boxes

print(greedy_load([[1, 3], [2, 2], [3, 1]], 4))
print(greedy_load([[5,10],[2,5],[4,7],[3,9]], 10))
print(greedy_load([[1,1]], 5))
print(greedy_load([[3,5],[2,5]], 5))
print(greedy_load([[1,10]], 1))
print(greedy_load([[2,3],[3,2],[1,5]], 100))
```

```python
# Problem Statement: At a lemonade stand, each lemonade costs exactly 5 dollars. Customers stand in a queue and purchase 1 lemonade each,
# presenting either a 5, ten, or twenty dollar bill. You start with absolutely 0 cash. Determine if you can provide
# mathematically perfect change to every single customer in the queue consecutively.
# Example Input: Bills [5, 5, 5, 10, 20]
# Example Output: True

def lemonade_change(bills):
    fives = 0
    tens = 0

    for bill in bills:

        if bill == 5:
            fives += 1

        elif bill == 10:
            if fives == 0:
                return False
            fives -= 1
            tens += 1

        elif bill == 20:
            if tens > 0 and fives > 0:
                tens -= 1
                fives -= 1
            elif fives >= 3:
                fives -= 3
            else:
                return False

    return True

print(lemonade_change([5, 5, 5, 10, 20]))
print(lemonade_change([5,5,10]))
print(lemonade_change([5,5,5,5,20,20]))
print(lemonade_change([5,10,10,5,20]))
```

```python
# Problem Statement: Given a matrix array of highly overlapping intervals, mathematically calculate the
# absolute minimum number of intervals required to be actively removed to completely eradicate all overlaps from the dataset.
# Example Input: [[1, 2], [2, 3], [3, 4], [1, 3]]
# Example Output: 1

def erase_overlap_intervals(intervals) :
  intervals.sort(key = lambda x: x[0])

  count = 0
  last_start, last_end = intervals[0]

  for start, end in intervals[1:] :
    if last_end > start :
      count += 1
    else:
      continue


  return count

print(erase_overlap_intervals([[1, 2], [2, 3], [3, 4], [1, 3]]))
print(erase_overlap_intervals([[1,2], [2,3], [3,4]]))
print(erase_overlap_intervals([[1,5], [2,6], [3,7], [4,8]]))
print(erase_overlap_intervals([[1,2], [1,2], [1,2]]))
print(erase_overlap_intervals([[5,10]]))
```

```python
# Problem Statement: You are provided an integer array where each specific element strictly represents the maximum
# physical jump length available from that exact position. Positi1d at the initial first index, definitively verify
# if it is mathematically possible to physically reach the absolute final index successfully.
# Example Input: [2, 3, 1, 1, 4]
# Example Output: True

def can_jump(nums):
    max_reach = 0

    for i, val in enumerate(nums):

        if i > max_reach:
            return False

        max_reach = max(max_reach, i + val)

        if max_reach >= len(nums) - 1:
            return True

    return True

print(can_jump([2,3,1,1,4]))
print(can_jump([3,2,1,0,4]))
print(can_jump([0]))
print(can_jump([1,0,0]))
print(can_jump([1,1,1,1]))
print(can_jump([5,0,0,0,0]))
print(can_jump([1,2,0,1]))
```

```python
# Minimum Number of Arrows to Burst Balloons
# Problem Statement: You must burst balloons deployed on a flat wall mapped via X coordinates.
# An arrow shot exactly vertically from an X position bursts all balloons spanning across that exact
# X coordinate horizontally. Mathematically determine the absolute minimum number of distinct arrows
# required to burst the entire balloon array.
# Example Input: [[10, 16], [2, 8], [1, 6], [7, 12]]
# Example Output: 2

def find_min_arrow_shots(ballons) :
  ballons.sort(key = lambda x: x[1])

  last_start, last_end = ballons[0]

  count = 1
  for start, end in ballons[1: ] :

    if start > last_end :
      count += 1
      last_end = end

  return ballons, count

print(find_min_arrow_shots([[10, 16], [2, 8], [1, 6], [7, 12]]))
print(find_min_arrow_shots([[1,2]]))
print(find_min_arrow_shots([[1,2], [3,4], [5,6], [7,8]]))
```
