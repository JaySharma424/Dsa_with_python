# Arrays vs. Linked Lists Comparison

When deciding between an Array and a Linked List, it is important to understand the technical and structural trade-offs. This table breaks down the core differences, making it easy to understand exactly when to use which data structure.

| Feature | Arrays | Linked Lists |
| :--- | :--- | :--- |
| **Memory Allocation** | **Contiguous:** Elements are stored in a single, continuous block of memory right next to each other. | **Non-Contiguous (Scattered):** Elements (nodes) are scattered throughout the memory heap and connected by pointers. |
| **Size** | **Fixed / Rigid:** You generally must declare the size upfront. Resizing requires creating a brand new, larger array and copying all data over. | **Dynamic / Flexible:** Can grow or shrink endlessly at runtime without ever needing to copy or shift data. |
| **Access Speed (Searching by Index)** | **Extremely Fast — O(1):** Supports "Random Access." The computer can mathematically calculate the exact location of any index instantly. | **Slow — O(N):** Only supports "Sequential Access." You must start at the Head and hop through pointers one by one to find an item. |
| **Insertion & Deletion** | **Slow — O(N):** Inserting or deleting an element at the beginning or middle forces you to physically shift every subsequent element over by one space. | **Extremely Fast — O(1):** If you already know the location, adding or removing a node simply requires rewiring two pointers. No shifting is required. |
| **Memory Overhead** | **Zero Overhead:** Every bit of allocated memory is strictly used to store the actual data. | **High Overhead:** Every single node requires extra memory to store the pointer(s) directing to the next/previous nodes. |
| **Cache Locality** | **Excellent:** Because data is stored physically side-by-side, the CPU can cache chunks of the array easily, making processing incredibly fast. | **Poor:** Because nodes are scattered randomly across the system's memory, the CPU cannot cache them efficiently, leading to micro-delays. |
| **Best Used When...** | You need to look up data frequently by its index, and the size of your dataset is relatively stable. | You need to constantly add and remove data dynamically, and you don't need to look up random elements by their index. |

* **Choose an Array** when you need to **read** data extremely fast (like looking up page 45 in a book).

* **Choose a Linked List** when you need to **write, add, or delete** data constantly without worrying about running out of space

**1. Singly Linked List**

**The Analogy**: You are given a clue that leads you to location A. At location A, you find a note that points you to location B. At location B, a final note points to location C.

**How it works**: Each piece of data (a "node") contains your item plus a single link pointing to the next node. You can only move in one direction—forward. If you are at B, you have no way of knowing what pointed to it.

```python
# create a singly linked list
class Node :
  def __init__(self, data) :
    self.data = data
    self.next = None
class SinglyLinkedList:
  def __init__(self) :
    self.head = None

  # Append a Node

  def append(self, data) :
    new_node = Node(data)
    if self.head is None :
      self.head = new_node
      return
    current = self.head

    while current.next is not None:
      current = current.next

    current.next = new_node

  # Add the node at the beginning of the linkedlist

  def prepend(self, value) :
    new_node = Node(value)
    new_node.next = self.head
    self.head = new_node


  # Delete Node
  def delete_node(self, target) :
    if not self.head :
      print("empty list")
      return

    prev = None
    temp = self.head
    while temp is not None and temp.data != target :
      prev = temp
      temp = temp.next

    if temp is None :
      print(' Not found ')
      return
    prev.next = temp.next
  # Delete Node from the beginning of linked list

  def delete_begining(self) :

    if not self.head:
      print("list is empty")
      return

    self.head = self.head.next

  # Display Node

  def display(self) :
    current = self.head
    while current is not None :
      print(current.data, end = '->')
      current = current.next

  # size of linked list

  def size(self) :
    count = 0
    if self.head is None :
      print('Empty linkedList')
      return
    current = self.head
    while current :
      count += 1
      current = current.next
    return count
list1 = SinglyLinkedList()
list1.append(1)
list1.append(2)
list1.append(3)
list1.append(4)
list1.append(5)
# list1.display()
# list1.delete_node(5)
# list1.size()

# list1.display()
# list1.prepend(10)
# list1.display()
# list1.delete_begining()
list1.display()
# list1.size()
```

```python
# Reverse a Singly Linked List
# Problem Statement: Assume you are provided the absolute Head pointer of a Singly Linked List.
# Reconstruct the entire structure so that all pointers violently change direction, completely reversing the list sequentially.
# Output the new Head pointer.
# Example Input: Head targeting [one -> two -> three -> null]
# Example Output: Head targeting [three -> two -> one -> null]



# Function to reverse the list
def reverse_list(head):
    previous = None
    current = head
    while current is not None:
        next_temp = current.next
        current.next = previous
        previous = current
        current = next_temp
    return previous

# Helper function to print linked list
def print_list(head):
    current = head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")

# Using the Node class defined in the previous cell
node1 = Node("one")
node2 = Node("two")
node3 = Node("three")
node1.next = node2
node2.next = node3

print("Original Linked List:")
print_list(node1)

# 2. Reverse the list
new_head = reverse_list(node1)

print("\nReversed Linked List:")
print_list(new_head)
```

```python
# Find Middle Node
# Problem Statement: Extract the mathematically central node of a given Singly Linked List.
# If the list possesses an explicitly even number of nodes, safely output the second middle node.
# Example Input: Head targeting [one -> two -> three -> four -> five]
# Example Output: Node(three)


# find the middle element or node
def middle_node(head) :
  slow = head
  fast = head
  while fast is not None and fast.next is not None :
    slow = slow.next
    fast = fast.next.next
  return slow

# print list
def print_list(head) :
  current = head
  while current is not None :
    print (current.data, end = ' ->')
    current = current.next


# creating node using Node class

node1 = Node('one')
node2 = Node('two')
node3 = Node('three')
node4 = Node('four')
node5 = Node('five')
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

print('Original Linked List :')
print_list(node1)
print()
print('Middle Node :')
print(middle_node(node1).data)
```

```python
# Merge Two Sorted Lists
# Problem Statement: Assume you possess two strictly ascending Singly Linked Lists.
# Weave them together securely to form one unified, perfectly sorted master list.
# Synthesize the new list by physically splicing the original nodes.
# Example Input: List A [one -> three], List B [two -> four]
# Example Output: [one -> two -> three -> four]


# ListNode class
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Merged sorted lists function
def merge_two_lists(list1, list2):
    dummy = ListNode(0)
    tail = dummy

    while list1 and list2:
        if list1.val < list2.val:
            tail.next = list1
            list1 = list1.next
        else:
            tail.next = list2
            list2 = list2.next
        tail = tail.next


    tail.next = list1 if list1 else list2

    return dummy.next


def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head


def print_list(head):
    result = []
    current = head
    while current:
        result.append(current.val)
        current = current.next
    return result


list1 = create_linked_list([1, 3])
list2 = create_linked_list([2, 4])

print("Input List 1:", print_list(list1))
print("Input List 2:", print_list(list2))

merged = merge_two_lists(list1, list2)
print("Merged:", print_list(merged))



```

```python
# Remove Nth Node From End
# Problem Statement: Assume a Singly Linked List and a positive integer K.
# Remove the mathematically precise K-th node from the absolute end of the list securely, and return the modified head pointer.
# Example Input: [1 -> 2 -> 3 -> 4 -> 5], K = 2
# Example Output: [1 -> 2 -> 3 -> 5]


# ListNode class define करो
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Remove Nth node from end function
def remove_nth_from_end(head, n):

    dummy = ListNode(0, head)
    left = dummy
    right = head


    for i in range(n):
        right = right.next


    while right:
        left = left.next
        right = right.next

    left.next = left.next.next

    return dummy.next

def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def print_list(head):
    result = []
    current = head
    while current:
        result.append(current.val)
        current = current.next
    return result



head = create_linked_list([1, 2, 3, 4, 5])
n = 2

print("Input List:", print_list(head))

head = create_linked_list([1, 2, 3, 4, 5])
result = remove_nth_from_end(head, n)
print("Output List:", print_list(result))
```
