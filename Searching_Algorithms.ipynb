### **Linear Search**

```python
# Given an Array and Target find the target in the given array and return the index if present otherwise return not found.
# If array is empty return -1
def linear_search(arr, target):
  n = len(arr)
  if n == 0:
    return -1
  for i in range(n):
    if arr[i] == target:
      return i
  return "Target Not found"

print(linear_search([34, 78, 12, 9, 89, 50, 41], 50))
print(linear_search([34, 78, 12, 9, 89, 50, 41], 41))
print(linear_search([34, 78, 12, 9, 89, 50, 41], 55))
print(linear_search([],2))
```

```python
# Write a Python function that uses a highly modified Linear Search to find the very last occurrence of a target element in a list containing duplicates.
# If it exists, return the specific index. If not, return -1.
# Example Input:
# arr = [10, 20, 30, 20, 40], target = 20
# Example Output: 3

def find_last_linearly(arr, target):
  n = len(arr)
  if n == 0 :
    return -1
  for i in range(n-1,-1, -1):
    if arr[i] == target:
      return i
  return "Not Found"

print(find_last_linearly([10, 20, 30, 20, 40], 20))
print(find_last_linearly([5, 8, 12, 7], 12))
print(find_last_linearly([1, 2, 3, 4], 9))
print(find_last_linearly([6, 6, 6, 6], 6))
```

```python
# You are analyzing raw sensor data that arrives continuously and remains completely unsorted.
# Write a Linear Search variant that scans the entire array to find and return the absolute maximum value present.
# Example Input:arr = [12, 5, 87, 22, 9]
# Example Output: 87
def maximun_element(arr):
  n = len(arr)
  if n == 0 :
    return None

  max_element = - float("inf")

  for i in range(n):
    if arr[i] > max_element:
      max_element = arr[i]
  return max_element

print(maximun_element([12, 5, 87, 22, 9]))
print(maximun_element([99, 4, 7, 2, 1]))
print(maximun_element([3, 8, 15, 20]))
print(maximun_element([-10, -3, -25, -1]))
print(maximun_element([]))
```

```python
# The Standard Search (First Occurrence) Given an unsorted array of integers, write a function to return the index
# of the first occurrence of a target integer. If not found, return -1.
# Input: arr = [4, 2, 7, 2, 9], target = 2
# Expected Output: 1
def first_occurance(arr, target):

  n = len(arr)

  if n == 0 :
    return []

  for i in range(n):
    if arr[i] == target:
      return i
  return -1


print(first_occurance([4, 2, 7, 2, 9], 2))
print(first_occurance([5, 8, 3, 1], 3))
print(first_occurance([10, 20, 30], 15))
print(first_occurance([7, 4, 5, 7], 7))
print(first_occurance([], 100))
```

```python
# 2.The Accumulator (Counting Occurrences) Given a list of student grades, count how many students scored exactly an 85.
# Input: grades = [85, 90, 85, 70, 85]
# Expected Output: 3

def counting_occurance(arr):
  n = len(arr)

  count = 0

  if n == 0:
    return []

  for i in range(n):
    if arr[i] == 85:
      count += 1
  if 85 not in arr:
    return "85 is not present"
  return count


print(counting_occurance([85, 90, 85, 70, 85]))
print(counting_occurance([60, 75, 85, 95]))
print(counting_occurance([70, 80, 90, 100]))
print(counting_occurance([85, 85, 85, 85]))
print(counting_occurance([]))
```

```python
# The Extremes (Finding Minimum) Given an array of stock prices for the week, find and return the minimum price.
# Input: prices = [120, 115, 130, 110, 125]
# Expected Output: 110

def find_minimum(arr):
  n = len(arr)
  if n == 0 :
    return None

  minimum_element = float('inf')

  for i in range(n) :
    if arr[i] < minimum_element :
      minimum_element = arr[i]
  return minimum_element


print(find_minimum([120, 115, 130, 110, 125]))
print(find_minimum([50, 60, 70, 80]))
print(find_minimum([90, 85, 88, 70]))
print(find_minimum([-5, -10, -3, -20]))
print(find_minimum([]))
```

```python
# Condition-Based Search (Filtering)Given a list of mixed integers,
#return a new list containing only the even numbers that are strictly greater than 10.
# Input: arr = [5, 12, 18, 9, 10, 22]
# Expected Output: [12, 18, 22]

def even_GreaterThan10(arr):

  n = len(arr)

  if n == 0 :
    return []

  result = []

  for i in range(n) :
    if arr[i] % 2 == 0 and arr[i] > 10 :
      result.append(arr[i])
  if len(result) == 0:
    return -1
  return result

print(even_GreaterThan10([5, 12, 18, 9, 10, 22]))
print(even_GreaterThan10([1, 3, 5, 7, 9]))
print(even_GreaterThan10([2, 4, 6, 8, 10]))
print(even_GreaterThan10([-12, 14, 20, -8, 11]))
print(even_GreaterThan10([12, 16, 24, 30]))
```

### **Binary Search**

```python
# Given an sorted array and a target. find the target in that array in O(log n) time complexity.
# Expected Input : [9, 12, 34, 41, 50, 78, 89]
# Expected Output: 5 ( 78 is at 5th index )

def Binary_search(arr, target):

  n = len(arr)

  if n == 0 :
    return []

  left, right = 0, n - 1

  while left <= right :

    mid = ( left + right ) // 2
    if arr[mid] == target :
      return mid
    elif arr[mid] > target:
      right = mid - 1
    else:
      left = mid + 1

  return "Target Not found"

print(Binary_search([9, 12, 34, 41, 50, 78, 89], 78))
print(Binary_search([9, 12, 34, 41, 50, 78, 89], 9))
print(Binary_search([9, 12, 34, 41, 50, 78, 89], 89))
print(Binary_search([9, 12, 34, 41, 50, 78, 89], 100))
print(Binary_search([25], 25))
```

```python
# Using Binary Search logic, find the index of the FIRST occurrence of a target element in a sorted array that contains duplicates.
# For example, if the array is [1, 2, 2, 2, 3, 4] and the target is 2,the algorithm should return index 1, not index 2 or 3.

def binary_first_occurance(arr, target):
  n = len(arr)
  if n == 0 :
    return []

  left , right = 0, n-1

  index = -1

  while left <= right :
    mid = ( left + right ) // 2

    if arr[mid] == target :
      index = mid
      right = mid - 1

    elif arr[mid] > target:
      right = mid
    else:
      left = mid + 1
  return index

print(binary_first_occurance([1, 2, 2, 2, 3, 4], 2))
print(binary_first_occurance([5, 5, 5, 7, 8, 9], 5))
print(binary_first_occurance([1, 3, 4, 6, 6, 6], 6))
print(binary_first_occurance([10, 20, 30, 40, 50], 30))
print(binary_first_occurance([2, 4, 6, 8, 10], 20))
```

```python
# Given a sorted array of distinct integers and a target value, return the exact index if the target is found.
# If the target is not found, return the precise index where it would be if it were inserted in order.
# You must write an algorithm with an O(log n) runtime complexity.Example Input:arr = [1, 3, 5, 6], target = 2
# Example Output:1

def insert_or_find_index(arr, target):

  n = len(arr)
  if n == 0 :
    return []

  left , right = 0, n - 1

  while left <= right :
    mid = ( left + right ) // 2

    if arr[mid] == target:
      return mid

    elif arr[mid] > target :
      right = mid - 1

    else:
      left = mid + 1

  return left


print(insert_or_find_index([1, 3, 5, 6], 2))
print(insert_or_find_index([1, 3, 5, 6], 5))
print(insert_or_find_index([2, 4, 6, 8, 10], 7))
print(insert_or_find_index([5, 7, 9], 9))
print(insert_or_find_index([2, 4, 6, 8, 10, 15, 17, 19, 20], 19))
```

```python
# Given a non-negative integer x, compute and return the square root of x. Return only the integer part (truncate any decimals) using Binary Search.
# Do not use built-in exponent functions like x ** 0.5 or math.sqrt.
# Example Input:x = 8
# Example Output:2

def find_sqrt(number):
  if number == 0 or number == 1 :
    return 1
  left , right = 1, number

  while left <= right:
    mid = ( left + right ) // 2

    if mid * mid == number :
      return mid
    elif mid * mid > number :
      right = mid - 1
    else:
      left = mid + 1
  return right

print(find_sqrt(8))
print(find_sqrt(9))
print(find_sqrt(25))
print(find_sqrt(27))
print(find_sqrt(1297))
```

```python
# Peak Index in a Mountain Array. An array is a mountain array if it strictly increases to a peak element and then strictly decreases.
# Given a mountain array, find the index of the peak element in O(log n) time complexity using Binary Search.
# Example Input:arr = [0, 10, 5, 2]
# Example Output:1 (The value 10 is the peak)

def found_mountain_element(arr):
  n = len(arr)
  if n == 0 :
    return []
  left, right = 0, n - 1

  while left < right:
    mid = (left + right) // 2
    if arr[mid] < arr[mid + 1] :
      left = mid + 1
    else:
      right = mid

  return left

print(found_mountain_element([0, 10, 5, 2]))
print(found_mountain_element([1, 3, 8, 12, 13, 4, 2]))
print(found_mountain_element([2, 5, 3]))
print(found_mountain_element([1, 4, 7, 16, 15, 9, 6, 2]))
print(found_mountain_element([3, 9, 8, 6, 1]))
print(found_mountain_element([4, 5, 6, 7, 0, 1, 2]))
```

```python
# Count Occurrences in a Sorted Array. Given a strictly sorted array and a target value, count exactly how many times the target appears.
# The algorithm must operate in O(log n) time.
# Example Input:arr = [1, 1, 2, 2, 2, 2, 3], target = 2
# Example Output:4
def count_occurance(arr, target):

  left , right = 0, len(arr) - 1

  first = -1

  while left <= right:

    mid = ( left + right ) // 2

    if arr[mid] == target:

      first = mid

      right = mid - 1

    elif arr[mid] < target:

      left = mid + 1

    else:

      right = mid - 1

  if first == -1 :

    return None

  left , right = 0, len(arr) - 1

  last = -1

  while left <= right:

    mid = ( left + right ) // 2

    if arr[mid] == target:

      last = mid

      left = mid + 1
    elif arr[mid] < target:

      left = mid + 1
    else:
      right = mid - 1

  return last - first + 1

print(count_occurance([1, 1, 2, 2, 2, 2, 3], 2))
print(count_occurance([1, 1, 1, 2, 3], 1))
print(count_occurance([0, 0, 1, 1, 2, 2, 3, 3, 3, 3, 4, 4, 5, 5], 5))
print(count_occurance([0, 0, 1, 1, 2, 2, 3, 3,], 0))
print(count_occurance([1,1,2,3,4,5,6,6], 0))
```

```python
# Find Minimum in Rotated Sorted Array. Suppose an array of unique integers sorted in ascending order is rotated at an unknown pivot.
# Find the minimum element in O(log n) time.
# Example Input:arr = [4, 5, 6, 7, 0, 1, 2]
# Example Output:0
def minimun_element_rotated(arr):
  n = len(arr)
  if n == 0 :
    return -1

  left, right = 0, n -1
  while left < right:
    mid = (left + right) // 2
    if arr[mid] > arr[right] :
      left = mid + 1
    else:
      right = mid
  return arr[left]

print(minimun_element_rotated([4, 5, 6, 7, 0, 1, 2]))
print(minimun_element_rotated([30, 40, 50, 10, 20]))
print(minimun_element_rotated([1, 2, 3, 4, 5, 6]))
print(minimun_element_rotated([15, 18, 2, 3, 6, 12]))
```

```python
# A sorted array is rotated at an unknown pivot (e.g., [4, 5, 6, 7, 0, 1, 2]). Search for a target value and return its index.
# Input: arr = [4, 5, 6, 7, 0, 1, 2], target = 0
# Expected Output: 4
def search_rotated(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid

        # Left half sorted
        if arr[left] <= arr[mid]:
            if arr[left] <= target < arr[mid]:
                right = mid - 1
            else:
                left = mid + 1

        # Right half sorted
        else:
            if arr[mid] < target <= arr[right]:
                left = mid + 1
            else:
                right = mid - 1


print(search_rotated([4, 5, 6, 7, 0, 1, 2], 0))
print(search_rotated([15, 18, 2, 3, 6, 12], 18))
print(search_rotated([30, 40, 50, 10, 20], 20))
print(search_rotated([1, 2, 3, 4, 5, 6, 7], 6))
print(search_rotated([6, 7, 8, 1, 2, 3, 4], 1))
```

```python
# Given a positive integer num, write a function which returns True if num is a perfect square, else False. Do not use built-in math libraries.
# Input: num = 16
# Expected Output: True
def is_square(n):
  if n == 0 :
    return n
  left, right = 0 , n
  while left <= right:
    mid = (left + right) // 2
    if mid * mid == n:
      return True
    elif mid * mid > n :
      right = mid - 1
    else:
      left = mid + 1
  return False

print(is_square(16))
print(is_square(25))
print(is_square(1))
print(is_square(0))
print(is_square(24))
```

### **Jump Search**

```python
import math

def jump_search(arr, target):

    n = len(arr)

    step = int(math.sqrt(n))

    prev = 0

    while arr[min(step, n) - 1] < target:

        prev = step

        step += int(math.sqrt(n))

        if prev >= n:

          return -1

    while arr[prev] < target:
      prev += 1

      if prev == min(step, n):

        return -1

      if arr[prev] == target:

        return prev

    return -1
print(jump_search([2,5,6,7,9,12], 10))
print(jump_search([2,5,6,7,9,12], 12))
```
