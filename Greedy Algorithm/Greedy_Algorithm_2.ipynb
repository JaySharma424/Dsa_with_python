```python
# Remove K Digits to Get the Smallest Possible Number
# Given a numeric string num and an integer k, remove exactly k digits from the string such that the resulting number is the smallest possible.
# Return the resulting number as a string. Leading zeros should be removed (unless the number is zero).
# num = "1432219", k = 3
# Output: 1219
# Explanation:Greedy removals minimize number

def remove_k_digits(num, k):
  stack = []

  for digit in num:
    while stack and k > 0 and stack[-1] > digit:
      stack.pop()
      k -= 1

    stack.append(digit)

  result = "".join(stack).lstrip("0")
  return result if result else "0"

print(remove_k_digits('1432219', 3))
print(remove_k_digits('10200', 1))
```

```python
# Maximum Number of Unique Substrings Using Greedy
# Given a string s, split it into the maximum number of substrings such that each substring is unique.
# You must split the string in order and cannot rearrange characters.
# Return the maximum number of unique substrings possible.
# s = "abac"
# Output: 3
# Explanation: Split as "a" | "b" | "ac"

def max_unique_substring(s):

  unique_values = set()

  count = 0

  i = 0

  while i < len(s) :

    j = i + 1

    while j <= len(s) :

      substring = s[i:j]

      if substring not in unique_values:

        unique_values.add(substring)

        count += 1

        i = j

        break

      j += 1

    else:
      break



  return count
print(max_unique_substring('abac'))
print(max_unique_substring('aba'))
print(max_unique_substring('abcd'))
print(max_unique_substring('aaaa'))
print(max_unique_substring('aab'))
print(max_unique_substring('abcabc'))
print(max_unique_substring('a'))
```

```python
# Problem Statement
# You are given an array prices where prices[i] is the price of a given stock on the i-th day.
# On each day, you may buy and/or sell the stock. You may hold at most one stock at a time — but you can buy and sell on the same day.
# Return the maximum profit you can achieve.
# input = [7, 1, 5, 3, 6, 4]

def max_profit(arr) :

  profit = 0

  n = len(arr)

  # i = 1

  for i in range(n-1):
    if arr[i] < arr[i + 1] :
      profit += arr[i + 1] -  arr[i]

  return profit


print(max_profit([7, 1, 5, 3, 6, 4]))
print(max_profit([1,2,3,4,5]))
print(max_profit([5,4,3,2,1]))
print(max_profit([3,3,5,0,0,3,1,4]))
print(max_profit([1,5,1,5]))
print(max_profit([2,2,2,2]))
```

```python
# Problem: Given an array of intervals, merge all overlapping intervals into one contiguous interval.
# Input: [[1, 3], [2, 6], [8, 10], [15, 18]]
# Output: [[1, 6], [8, 10], [15, 18]]

def merge_intervals(nums) :
  nums.sort(key = lambda x : x[0])

  result = [nums[0]]
  for start, end in nums :
    last_start, last_end = result[-1]
    if last_end > start :
      result[-1][1] = max(last_end, end)
    else:
      result.append([start, end])
  return result

print(merge_intervals([[1, 3], [2, 6], [8, 10], [15, 18]]))
```

```python
# Problem: Given an array of meeting time intervals, find the absolute minimum number of conference rooms required.
# Input: [[0, 30], [5, 10], [15, 20]]
# Output: 2

def meeting_room(arr):
  arr.sort(key = lambda x: x[0])

  temp = [arr[0]]
  count = 1
  i = 0

  for start, end in arr[1:]:
    last_start, last_end = temp[-1]
    if last_end > start:
      count += 1
      i += 1
      temp = [arr[i]]
  return count


print(meeting_room([[0, 30], [5, 10], [15, 20], [2, 7], [2, 25]]))
print(meeting_room([[0, 30], [5, 10], [15, 20]]))
print(meeting_room([[1,5],[2,6],[3,7]]))
```

```python
# Problem: Insert a new interval into a sorted array of non-overlapping intervals and merge if necessary.
# Input: intervals = [[1, 3], [6, 9]], newInterval = [2, 5]
# Output: [[1, 5], [6, 9]]

def insert_interval(intervals, newInterval):
  intervals.append(newInterval)
  intervals.sort(key = lambda x: x[0])

  result = []

  for interval in intervals:
    if not result or result[-1][1] < interval[0] :
      result.append(interval)
    else:
      result[-1][1] = max(result[-1][1], interval[1])

  return result
print(insert_interval([[1, 3], [6, 9]], [2,5]))
print(insert_interval([[3, 5], [7, 9]], [1, 2]))
print(insert_interval([[1, 2], [3, 5]], [6, 8]))
print(insert_interval([[1, 2], [3, 5], [6, 7], [8, 10], [12, 16]], [4, 9]))
```

```python
# IPO (Maximize Capital)
# Problem: You have initial capital k and arrays of profits and capital required for various projects.
# Select up to n projects to maximize your final capital.
# Input: k = 2, w = 0, profits = [1, 2, 3], capital = [0, 1, 1] Output: 4
# Greedy Logic: Push all projects you can currently afford into a Max-Heap based on their profit.
# Greedily pop the most profitable project, add its profit to your capital,
# and then check if this new capital allows you to afford new projects. Repeat up to n times.

import heapq

def find_maximized_capital(n, w, profits, capital):
    # Min heap — [capital_required, profit] — saste projects pehle
    min_heap = sorted(zip(capital, profits))  # capital se sort

    # Max heap — sirf profits (negative kyunki python max heap nahi hai)
    max_heap = []

    for _ in range(n):
        # Step 1: jo projects afford ho sakein — max heap mein daalo
        while min_heap and min_heap[0][0] <= w:
            cap, pro = heapq.heappop(min_heap)
            heapq.heappush(max_heap, -pro)   # negative — max heap trick

        # Step 2: koi affordable project nahi — band karo
        if not max_heap:
            break

        # Step 3: sabse zyada profitable project lo
        w += -heapq.heappop(max_heap)        # negative wapas positive

    return w


print(find_maximized_capital(2, 0, [1, 2, 3], [0, 1, 1]))
```
