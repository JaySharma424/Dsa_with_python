# Introduction to Recursion

The Unix operating system uses a recursive directory structure to organize files and folders. Each directory in the file system can contain sub-directories, which can, in turn, contain more sub-directories, and so on. This structure can be navigated using a recursive algorithm, which starts at the root directory and then recursively navigates through each sub-directory until the desired file or folder is found. This approach allows for a flexible and scalable file system that can handle many files and folders.

Recursion can be used in many other ways in programming. Another example is the creation of fractals, which are complex geometric patterns that repeat themselves at different scales. A fractal can be created using a recursive algorithm that repeatedly applies rules to a simple geometric shape, such as a line or a triangle. As the algorithm runs, the shape is repeated and scaled down, creating a complex pattern that exhibits self-similarity. This technique is used in many computer graphics and animation and can produce stunningly beautiful visual effects.

## Recursion Concept
Recursion is a technique in which a function calls itself until it reaches a base case. The base case is a condition that stops the recursive function from calling itself. Each recursive call creates a new instance of the function on the call stack, meaning it has its own set of variables and parameters.

### How to use Recursion?
To use recursion, you must define a function that calls itself. The function should have a **base case** that stops the recursion and a **recursive case** that calls itself with a modified input.

```python
# Write a python programme to print factorial of a number using recursion

def factorial_recursion(number):
  if number == 0 :
    return 0
  if number == 1 :
    return 1
  return number * factorial_recursion(number - 1)
print(factorial_recursion(5))
print(factorial_recursion(6))
print(factorial_recursion(7))
```

```python
# Define a function to reverse a string of text utilizing call stack memory

def reverse_string_recursion(text) :

  n = len(text)

  if n <= 1 :
    return text.upper()

  return reverse_string_recursion(text[1::]) + text[0].lower()

print(reverse_string_recursion('Hello'))
print(reverse_string_recursion('Dhananjay'))
print(reverse_string_recursion(''))
```

```python
# Define a function to print fabonacci number using multiple recursion.

def fab_multi_recursion( n ) :
  if n == 0 :
    return 0
  if n == 1:
    return n
  return fab_multi_recursion(n -1) + fab_multi_recursion(n - 2)
print(fab_multi_recursion(5))
print(fab_multi_recursion(6))
print(fab_multi_recursion(1))
```

```python
# Write a recursive function that calculates the sum of all natural numbers up to
# and including a given positive integer.Example Input: five Example Output: fifteen

def natural_sum_recursion(n) :
  if n <= 0 :
    return 0
  if n == 1:
    return n
  return natural_sum_recursion(n-1) + n
print(natural_sum_recursion(5))
print(natural_sum_recursion(10))
print(natural_sum_recursion(2))
```

```python
# Problem Statement: Write a recursive function that accepts a base number
# and an exponent, and mathematically returns the base raised to the power of the exponent.
# Example Input: base two, exponent three
# Example Output: eight

def power_cal_recursion(base, exponent) :
  if base <= 0 :
    return 0
  if exponent == 0:
    return 1
  return base * power_cal_recursion(base, exponent - 1)
print(power_cal_recursion(2,10))
```

```python
# Problem Statement: Write a recursive function that accepts an array and a target value,
# mathematically returning the absolute frequency of the target value occurring within the array boundaries.
# Example Input: array [one, two, two, three], target two
# Example Output: two

def fre_count_recursion(arr, i, target) :
  # count = 0
  n = len(arr)
  if i >= n :
    return 0
  if arr[i] == target:
    # count += 1
    return 1 + fre_count_recursion(arr, i + 1, target)
  else:
    return fre_count_recursion(arr, i + 1, target)

print(fre_count_recursion([5,5,5,51,1,1,2,2,2,2,4], 0, 5))
print(fre_count_recursion(["one", "two", "two", "three"], 0, "two"))
```

```python
# Problem Statement: Write a recursive function that accepts a string of text and
# returns a deeply nested boolean indicating whether the string reads identically forwards and backwards.
# Example Input: 'racecar'
# Example Output: True

def is_palindrome_recursion(text, i) :

  n = len(text)
  if i == n // 2 :
    return True
  if text[i].lower() != text[n - i -1].lower() :
    return False
  else:
    return is_palindrome_recursion(text, i + 1)



print(is_palindrome_recursion("racecar", 0))
print(is_palindrome_recursion('Nitin', 0))
print(is_palindrome_recursion('a;aldkjf;lkadjf;o', 0))
```

```python
# Problem Statement: Write a recursive function that reverses an array in-place without
# utilizing any external array structures or looping constructs.
# Example Input: ['one', 'two', 'three', 'four']
# Example Output: ['four', 'three', 'two', 'one']

def reverse_array_recursion(arr, i, j) :

  if i >= j:
    return

  arr[i] , arr[j] = arr[j] ,arr[i]

  reverse_array_recursion(arr, i + 1, j - 1)
  return arr


arr = ['one', 'two', 'three', 'four']
print(reverse_array_recursion(arr, 0 , len(arr)-1))
# reverse_array_recursion([1, 2, 3, 4, 5], 4)
```

```python
# Problem Statement: Write an advanced multiple recursion algorithm that generates and
# returns an array containing all mathematical permutations of a provided string.
# Example Input: 'abc'
# Example Output: ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']

def generate_permutations(text):
  if len(text) <= 1:
    return [text]
  collections = []

  for i in range(len(text)) :
    anchor = text[i]
    remaining = text[:i] + text[i + 1:]
    for sub_permutation in generate_permutations(remaining) :
      collections.append(anchor + sub_permutation)
  return collections
print(generate_permutations('abc'))
print(generate_permutations('xyz'))
print(generate_permutations('123'))
```

```python
# find the sum of digit using recursion.
def digit_sum_recursion(number) :
  if number < 10 :
    return number
  return (number % 10) + digit_sum_recursion(number // 10)
print(digit_sum_recursion(525))
```

```python
# Given a number find that the number is perfect square or not. If yes, return True otherwise return False.
def is_perfect_square(n, mid):
  if mid <= 1:
    return False
  if mid * mid == n :
    return True
  elif mid * mid > n:
    return is_perfect_square(n, mid - 1)
  else:
    return False

n = 729
print(is_perfect_square(n, n // 2))
n = 730
print(is_perfect_square(n, n // 2))
```

```python
# Given a number find that the number is perfect Cube or not. If yes, return True otherwise return False.
def is_perfect_cube(n, low, high) :
  if low > high :
    return False
  mid = (low + high) // 2

  if mid ** 3 == n:
    return True
  elif mid ** 3 > n :
    return is_perfect_cube(n, low, mid -1)
  elif mid ** 3 < n:
    return is_perfect_cube(n, mid + 1, n)


print(is_perfect_cube(27, 0, 27))
print(is_perfect_cube(216, 0, 216))
print(is_perfect_cube(343, 0, 343))
print(is_perfect_cube(1330, 0, 1330))
```

```python
# (Flatten a Nested Dictionary): Write a recursive function that deep-traverses
# a nested string-to-object dictionary and flattens it into a single-level layout.
# Combine matching child hierarchies using dot notation (parent.child).
def flatten(nested_dict, prefix = '', result = {}) :
  for key , value in nested_dict.items() :
    print('key', key)
    print('value', value)
    print('prefix', prefix)
    full_key = prefix + '.' + key if prefix else key
    print('full_key', full_key)

    if isinstance(value, dict) :
      print('recursion reached ')
      flatten(value, full_key, result)
    else:
      result[full_key] = value
      print('result', result)
  return result

print(flatten({"name":"Dhananjay", "address":{"city":"Patna","pin":{"code":"800001"}}}, ' ', {}))

```

```python
# (Balanced Parentheses Generation): Write an execution tree algorithm
# that accepts an integer n pairs of brackets and generates all valid configurations
# of perfectly balanced, well-formed open and closed parentheses strings.

def generate_balanced_paren(n) :
  result = []
  def recursion(curr, open_count, close_count) :
    if len(curr) == 2 *n :
      result.append(curr)
      return
    if open_count < n :
      recursion(curr + "(" , open_count + 1, close_count)
    if close_count < open_count :
      recursion(curr + ")", open_count , close_count + 1)
  recursion("", 0, 0)
  return result
print(generate_balanced_paren(2))
print(generate_balanced_paren(3))
```

```python
# (The Tower of Hanoi):** Devise a multiple-recursive mapping breakdown that prints
#  the perfect sequence of minimal disc movements required to transfer `n` nested layers
# from a source rod to a destination rod using an auxiliary buffer tower.
def hanoi(n, source, destination, auxiliary):
    # Base Case: Agar koi dabba bacha hi nahi (n == 0), to ruk jao.
    if n == 0:
        return

    # STEP 1: Upar ke (n-1) dappon ko Source se uthakar Auxiliary par daalo
    # Is baar destination ka kaam 'auxiliary' rod karegi.
    hanoi(n - 1, source, auxiliary, destination)

    # STEP 2: Sabse bade (n-th) dabbe ko Source se Destination par move karo
    print(f"Move disc {n} from {source} -> {destination}")

    # STEP 3: Jo (n-1) dappe Auxiliary par the, unhe uthakar Destination par daalo
    # Is baar source ka kaam 'auxiliary' rod karegi.
    hanoi(n - 1, auxiliary, destination, source)
hanoi(3, [], [], [])
```

```python
def hanoi(n, source, destination, auxiliary):
    """
    Tower of Hanoi ka recursive function jo movements print karta hai.
    """
    # Base Case: Agar koi dabba nahi bacha, toh ruk jao
    if n == 0:
        return

    # Step 1: Upar ke n-1 dappon ko Source se Auxiliary par shift karo
    hanoi(n - 1, source, auxiliary, destination)

    # Step 2: Sabse bade dabbe ko Source se Destination par rakho
    print(f"Move disc {n} from {source} -> {destination}")

    # Step 3: Jo n-1 dappe Auxiliary par the, unhe Destination par shift karo
    hanoi(n - 1, auxiliary, destination, source)


# --- MAIN PROGRAM WITH FIXED INPUT ---
if __name__ == "__main__":
    print("=== TOWER OF HANOI SOLVER ===")

    # Humne input value yahan fix kar di hai (3 discs)
    total_discs = 3

    # Formula: (2^n) - 1 moves lagenge
    total_moves = (2 ** total_discs) - 1

    print(f"Humne total discs select kiye hain: {total_discs}")
    print(f"Iske liye sabse kam {total_moves} steps lagenge.\n")
    print("--- Kadam-by-Kadam Steps ---")

    # Function Call:
    # 'A' = Pehli Table (Source)
    # 'C' = Aakhiri Table (Destination)
    # 'B' = Beech wali Table (Helper/Auxiliary)
    hanoi(total_discs, source='A', destination='C', auxiliary='B')

    print("\nBadhai ho! Saare dappe Table C par safely pahunch gaye. 🎉")
```
