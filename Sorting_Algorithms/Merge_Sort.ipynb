### **Merge Sort**

Best Case: O(n log n).

Average Case: O(n log n).

Worst Case: O(n log n).

```python
# Given an unsorted array sort the array using merge sort.

def merge_sort(arr) :
  n = len(arr)
  if n == 0 :
    return -1
  if n == 1:
    return arr

  mid = len(arr) // 2
  left = arr[:mid]
  right = arr[mid:]

  left_sort = merge_sort(left)
  right_sort = merge_sort(right)

  return merge(left_sort, right_sort)
def merge(left, right):
  sorted = []
  i = j = 0
  while i < len(left) and j < len(right) :
    if left[i] <= right[j] :
      sorted.append(left[i])
      i += 1
    else:
      sorted.append(right[j])
      j += 1
  while i < len(left) :
    sorted.append(left[i])
    i += 1
  while j < len(right) :
    sorted.append(right[j])
    j += 1
  return sorted

print(merge_sort([38, 27, 43, 3, 9, 82, 10]))
print(merge_sort([1, 2, 3, 4]))
print(merge_sort([5, 1, 5, 2]))
print(merge_sort([10, 7, 8, 9, 1, 5]))
```

```python
# Implement Merge Sort to handle a completely reverse-ordered array of integers correctly.
# Example Input: [9, 8, 7, 6, 5]
# Example Output: [5, 6, 7, 8, 9]
def merge_reverse_ordered(arr) :
  n = len(arr)
  if n == 0 :
    return -1
  if n == 1:
    return arr
  mid = n // 2
  left_half = arr[:mid]
  right_half = arr[mid: ]
  left_sort = merge_reverse_ordered(left_half)
  right_sort = merge_reverse_ordered(right_half)
  return merge(left_sort, right_sort)


def merge(left, right) :
  i = j = 0
  sort = []
  while i < len(left) and j < len(right) :
    if left[i] > right[j]:
      sort.append(right[j])
      j += 1
    else:
      sort.append(left[i])
      i += 1
  while j < len(right):
    sort.append(right[j])
    j += 1
  while i < len(left) :
    sort.append(left[i])
    i += 1
  return sort
print(merge_reverse_ordered([9, 8, 7, 6, 5]))
print(merge_reverse_ordered([5, 4, 3, 2, 1]))
print(merge_reverse_ordered([5]))
```

```python
# Use Merge Sort to sort a list of dictionaries based on a specific key value ('score').
# This simulates how databases sort JSON records.
# Example Input: [{'id': 1, 'score': 88}, {'id': 2, 'score': 95}, {'id': 3, 'score': 72}]
# Example Output: [{'id': 3, 'score': 72}, {'id': 1, 'score': 88}, {'id': 2, 'score': 95}]

def merge_sort_dict(arr) :
  n = len(arr)
  if n == 0 :
    return -1
  if n == 1 :
    return arr
  mid = n // 2
  left = arr[ :mid]
  right = arr[mid: ]
  lt_sort = merge_sort_dict(left)
  rt_sort = merge_sort_dict(right)
  return merge(lt_sort, rt_sort)
def merge(left, right) :
  sort = []
  i = j = 0
  while i < len(left) and j < len(right) :
    if left[i]['score'] < right[j]['score'] :
      sort.append(left[i])
      i += 1
    else:
      sort.append(right[j])
      j += 1
  while i < len(left) :
    sort.append(left[i])
    i += 1
  while j < len(right) :
    sort.append(right[j])
    j += 1
  return sort

print(merge_sort_dict([{'id': 1, 'score': 88}, {'id': 2, 'score': 95}, {'id': 3, 'score': 72}]))
```

```python
# Count Inversions in an Array**Find how many pairs (i, j) exist such that i < j and `arr[i] > arr[j]`.

def count_inversions(arr):
    if len(arr) <= 1:
        return arr,  0           # base case — 1 element, 0 inversions

    mid = len(arr) // 2

    # Division phase — recursive calls
    left, left_count  = count_inversions(arr[:mid])
    right, right_count = count_inversions(arr[mid:])

    # Merge phase — yahan inversions count honge
    sorted_arr = []
    inversions = left_count + right_count   # subproblems ke inversions add karo
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            sorted_arr.append(left[i])
            i += 1
        else:
            # KEY LINE: left[i] se lekar left ke end tak sab inversions hain
            inversions += len(left) - i
            sorted_arr.append(right[j])
            j += 1

    # Remaining elements append karo (inversion nahi, sirf leftover)
    sorted_arr.extend(left[i:])
    sorted_arr.extend(right[j:])

    return sorted_arr, inversions
sorted_arr, count = count_inversions([8, 4, 2, 9, 3])
print("Inversions:", count)    # 6
print("Sorted:", sorted_arr)   # [2, 3, 4, 8, 9]
print(count_inversions([2, 4, 1, 3, 5]))
print(count_inversions([8, 4, 2, 9, 3]))
```

```python
# Reverse Pairs** Find the number of important reverse pairs where i < j and `arr[i] > 2 * arr[j]`.*
# Use Merge Sort but add a separate counting step before the standard merge logic to evaluate the `2 *` condition.
def count_reverse_pair(arr) :
  n = len(arr)
  if n == 0 :
    return [] , 0
  if n == 1 :
    return arr, 0

  mid = n // 2
  lt, lc = count_reverse_pair(arr[ :mid])
  rt, rc = count_reverse_pair(arr[mid: ])

  count = lc + rc

  j = 0

  for i in range(len(lt)) :
    while j < len(rt) and lt[i] > 2 * rt[j] :
      j += 1
    count += j

  i = k = 0
  result = []

  while i < len(lt) and k < len(rt) :
    if lt[i] < rt[k] :
      result.append(lt[i])
      i += 1
    else:
      result.append(rt[k])
      k += 1
  result.extend(lt[i:])
  result.extend(rt[k:])
  return result, count

print(count_reverse_pair([8, 4, 2, 9, 3]))
```

```python
# Merge Two Sorted Arrays In-Place**
# * **Problem:** Given two sorted arrays, merge them into one sorted sequence without using extra space.

def merge_two_array_inPlace(arr1, arr2) :
  if len(arr1) == 0 and len(arr2) == 0 :
    return -1
  i = len(arr1) - 1
  j = 0
  while i >= 0 and j < len(arr2) :
    if arr1[i] >= arr2[j] :
      arr1[i] , arr2[j] = arr2[j] , arr1[i]
      arr1.sort()
      arr2.sort()
    else:
      break
  return arr1 + arr2
arr1 = [1, 3, 5, 7]
arr2 = [2, 4, 6, 8]
print(merge_two_array_inPlace(arr1, arr2))
```
