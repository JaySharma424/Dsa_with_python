### **Bubble Sort**

```python
# Given an array and sort the array using Bubble sort T.C --> O(n2)
# [64, 34, 12, 25, 4, 2]

def bubble_sort(arr):

  n = len(arr)
  if n == 0 :
    return - 1
  if n == 1:
    return arr

  for i in range(n):

    swapped = False

    for j in range(n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j +1] = arr[j+1] , arr[j]
        swapped = True

    print(arr)

    if not swapped:
      break

  return arr

print(bubble_sort([64, 1, 35, 25, 0, 2]))
```

```python
# Write a Python function that uses Bubble Sort to sort a list of numbers in strictly descending order rather than ascending order.
# Example Input[3, 1, 4, 1, 5, 9]
# Example Output[9, 5, 4, 3, 1, 1]

def descending_bubble_sort(arr):
  n = len(arr)

  if n == 0:
    return -1
  if n == 1:
    return arr

  sorted = 1

  for i in range(n) :

    swapped = False

    for j in range(n - i - 1):
      if arr[j] < arr[j + 1] :
        arr[j] , arr[j + 1] = arr[j + 1], arr[j]
        swapped = True
        sorted = 0

    if not swapped :
      break

  if sorted == 1:
    return "Already Sorted"

  return arr


print(descending_bubble_sort([3, 1, 4, 1, 5, 9]))
print(descending_bubble_sort([9,8,6,5,3,1,0]))
print(descending_bubble_sort([1,6,2,12,3,9,5]))
```

```python
# **3. Sort an Array of Strings by Length**
# * **Problem:** Given an array of strings, sort them so that shorter strings appear first.


def sort_by_string_length(arr) :

  n = len(arr)
  if n == 0 :
    return -1
  if n == 1:
    return arr

  for i in range(n):
    swapped = False

    for j in range(n - i - 1) :

      if len(arr[j]) > len(arr[j + 1]) :

        arr[j], arr[j + 1] = arr[j + 1], arr[j]

        swapped = True
    if not swapped :
      break
  return arr


print(sort_by_string_length(["hello", "Aarushsingh" ,"Ram", "romeao", "Dhananjay"]))
print(sort_by_string_length(["rohit"]))
```

```python
# Check if an Array is Sorted in One Pass**
# Write a function that returns `True` if an array is sorted, using the mechanics of Bubble Sort.
# Run exactly one pass of the inner loop. If no swaps occur, return `True`. executes in $O(n)$ time.

def check_array_in_one_pass(arr):

  # Assuming the array is not empty

  for i in range(len(arr)) :
    swapped = False

    for j in range(len(arr) - i - 1):
      if arr[j] > arr[j + 1]:
        swapped = True
        break

    if swapped :
      return False
    else:
      return True
  # return swapped


print(check_array_in_one_pass([1, 2, 3, 4, 5]))
print(check_array_in_one_pass([1, 3, 2, 4, 5]))
print(check_array_in_one_pass([7, 7, 7, 7]))
```

```python
# **5. Sort by Absolute Difference from a Target Value X**
# * **Problem:** Given an array and a value X, sort the array based on the absolute difference between the elements and X.


def Absolute_difference_sort(arr, x):

  # assuming array is not empty

  for i in range(len(arr)) :

    for j in range(len(arr) - i - 1) :

      if abs(arr[j] - x) > abs(arr[j + 1] - x) :
        arr[j] , arr[j + 1] =  arr[j + 1] , arr[j]

  return arr

print(Absolute_difference_sort([10, 5, 3, 9, 2], 7))
print(Absolute_difference_sort([4, 1, 8, 6, 3], 4))
print(Absolute_difference_sort([-5, -2, 0, 3, 7], 1))
```

### **Selection Sort**

```python
# Given an array sort the array using selection sort. if array is is_sorted return 'already is_sorted'.

def standard_selection_sort(arr):
    n = len(arr)
    is_sorted = True

    for i in range(n - 1):
        min_index = i

        for j in range(i + 1, n):
            if arr[j] <= arr[min_index]:
                min_index = j

        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]
            is_sorted = False
        print(arr)

        if is_sorted:
          return "Already sorted"

    return arr


print(standard_selection_sort([25, 27, 6, 9, 2]))
print(standard_selection_sort([25, 27, 32, 65, 99]))
print(standard_selection_sort([29, 10, 14, 37, 13]))
print(standard_selection_sort([5, 2, 5, 1, 2]))
print(standard_selection_sort([1, 2, 3, 4, 5]))
```

```python
# 1.Find the Kth Smallest Element
# * **Problem:** Find the Kth smallest element in an unsorted array without fully sorting it.

def kth_smallest_element(arr, k):
  if k == 0 or len(arr) < k :
    return -1

  for i in range(k) :

    smallest = i

    for j in range(1+i , len(arr)):

      if arr[j] <= arr[smallest] :

        smallest = j

    if smallest != i :

      arr[i], arr[smallest] = arr[smallest], arr[i]

  return arr[k-1]  # kth element will be returned

print(kth_smallest_element([25, 27, 6, 9, 2], 2))
print(kth_smallest_element([5, 2, 5, 1, 2], 3))
print(kth_smallest_element([1, 2, 3, 4, 5], 6))
```

```python
# **2. Stable Selection Sort*** Standard Selection Sort is unstable. Modify it so that identical elements
# maintain their relative initial order.

def stable_selection_sort(arr) :

  n = len(arr)


  for i in range(n-1) :

    smallest = i

    for j in range(i + 1 ,n) :
      if arr[j] < arr[smallest] :
        smallest = j

    smallest_element = arr[smallest]

    while smallest > i:
      arr[smallest] = arr[smallest-1]
      smallest -= 1

    arr[i] = smallest_element

  return arr

print(stable_selection_sort([25, 27, 6, 9, 2]))
print(stable_selection_sort([5, 2, 5, 1, 2]))
```

```python
# **4. Maximize Array Sum After K Negations**
# * **Problem:** Given an array, negate exactly K elements to maximize the total sum of the array.

def Maximize_array_sum_after_negations(arr, k):
    n = len(arr)
    if n == 0:
        return -1

    # Selection sort in ascending order
    for i in range(n - 1):
        smallest = i
        for j in range(i + 1, n):
            if arr[j] < arr[smallest]:
                smallest = j
        if smallest != i:
            arr[i], arr[smallest] = arr[smallest], arr[i]

    # Negate negative numbers first
    for i in range(n):
        if k > 0 and arr[i] < 0:
            arr[i] = -arr[i]
            k -= 1

    # If k is still odd, flip the element with minimum absolute value
    if k % 2 == 1:
        min_index = 0
        for i in range(1, n):
            if abs(arr[i]) < abs(arr[min_index]):
                min_index = i
        arr[min_index] = -arr[min_index]

    return sum(arr)

print(Maximize_array_sum_after_negations( [4, 2, 3], 1)) # 5
print(Maximize_array_sum_after_negations([-8, -3, -5, -2], 6)) # 18
print(Maximize_array_sum_after_negations( [3, -1, 0, 2], 3)) # 6
print(Maximize_array_sum_after_negations([2, -3, -1, 5, -4], 2)) # 13
print(Maximize_array_sum_after_negations([1, 3, 2, 4], 1)) # 8
print(Maximize_array_sum_after_negations([-2, 0, 5, -1], 4)) # 8
```

### **Insertion Sort**

```python
# write a python funtion to sort an array by using insertion sort.
def insertion_sort(arr):
    # Start from the second element (index 1), assuming the first is already sorted
    for i in range(1, len(arr)):
        # Store the current element to be positioned securely in a variable
        key = arr[i]

        # Initialize a pointer 'j' for tracking the sorted portion of the array
        # It starts exactly one position to the left of the current element
        j = i - 1

        # Move elements of the sorted portion that are strictly greater than the key
        # one physical position ahead of their current position to create a gap
        while j >= 0 and key < arr[j]:
            # Shift the larger element to the right
            arr[j + 1] = arr[j]
            # Decrement the pointer to check the next element to the left
            j -= 1

        # Place the stored key into its correct sorted location (the gap we created)
        arr[j + 1] = key

        print(arr)

    # Return the fully sorted array
    return arr
print(insertion_sort([12, 11, 13, 5, 6]))
print(insertion_sort([1, 2, 3, 5, 4]))
print(insertion_sort([2,25,7,8,12,9]))
```

```python
# Given a list of tuples representing students and their ages e.g., (Name, Age),
#use Insertion Sort to arrange the list strictly by Age in ascending order.
# Example Input: [('John', 25), ('Alice', 20), ('Bob', 30)]
# Example Output: [('Alice', 20), ('John', 25), ('Bob', 30)]


def sort_tuples(arr):
  n = len(arr)
  if n == 0 :
    return -1

  for i in range(1, n) :
    key = arr[i]
    j = i - 1

    while j >= 0 and key[1] < arr[j][1] :
      arr[j + 1] = arr[j]

      j -= 1
    arr[j + 1] = key

  return arr


print(sort_tuples([('John', 25), ('Alice',20), ('Bob',30)]))
print(sort_tuples([('Alice', 20), ('John', 25), ('Bob', 30)]))
print(sort_tuples([('Bob', 30), ('John', 25), ('Alice', 20)]))
print(sort_tuples([('John', 25)]))
print(sort_tuples([]))
```

```python
# **1. Binary Insertion Sort** Improve the standard Insertion Sort algorithm to reduce the number of comparisons.
# using Binary Search on the already sorted portion of the array to find the correct index in $O(\log n)$ comparisons.

def binary_insertion_sort(arr) :
  n = len(arr)
  if n == 0 :
    return -1
  for i in range(1, len(arr)):
    key = arr[i]

    # Binary search
    low, high = 0, i

    while low < high:
      mid = (low + high) // 2
      if key <= arr[mid] :
        high = mid
      else:
        low = mid + 1

    # Elements ko right shift karo [lo..i-1]
    for j in range(i, low ,  -1) :
      arr[j] = arr[j - 1]
    arr[low] = key

  return arr

print(binary_insertion_sort([12, 11, 13, 5, 6]))
print(binary_insertion_sort([1, 2, 3, 5, 4]))
print(binary_insertion_sort([2,25,7,8,12,9]))
print(binary_insertion_sort([1, 2, 3, 4, 5, 6]))
print(binary_insertion_sort([4, 3, 2, 1]))
```

```python
# **2. Sort a Nearly Sorted Array (K-Sorted Array)**
# * **Problem:** Sort an array where each element is at most K places away from its target position.

def k_times_sorted(arr, k) :
  n = len(arr)
  if n == 0 :
    return -1

  for i in range(1, n) :
    key = arr[i]
    j = i - 1

    while j >= 0 and i - j <= k and arr[j] > key:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j+1] = key
  return arr

print(k_times_sorted( [3, 1, 2, 5, 4, 6], 2))
print(k_times_sorted([1, 2, 3, 4, 5], 2))
print(k_times_sorted([2, 1, 3, 5, 4, 6], 1))
print(k_times_sorted([-1, -3, -2, 0, 2], 3))
print(k_times_sorted([3, 1, 2, 2, 4], 2))
```

```python
# **4. Running Median of a Data Stream** Maintain and return the median of a stream of incoming integers dynamically.
def running_median(stream) :
  n = len(stream)
  if n == 0 :
    return -1
  sorted_arr = []
  median = []

  for num in stream :
    sorted_arr.append(num)
    i = len(sorted_arr) - 2

    while i >= 0 and sorted_arr[i] > num :
      sorted_arr[i + 1] = sorted_arr[i]

      i -= 1
    sorted_arr[i + 1] = num

    left, right = 0, len(sorted_arr) - 1
    mid = (left + right) // 2

    if len(sorted_arr) % 2 == 0 :

      running_median = ((sorted_arr[mid]+sorted_arr[mid + 1]) //2)

    else:
      running_median = sorted_arr[mid]
    median.append(running_median)
  return median



print(running_median([5, 15, 1, 3]))
```
